```python
#importing libaries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
```

```python
#Display the first five rows of the data.
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
#For each category in the cyl column, calculate the average mpg for all cars with that cyl value.
df.groupby('cyl')['mpg'].mean()
```

```python
#Create a histogram using the mpg column.
import matplotlib.pyplot as plt
plt.hist(df['mpg'], edgecolor = 'black', linewidth = 1)
plt.xlabel('Miles per Gallon')
plt.ylabel('Frequency')
plt.title('Histogram of mpg')
plt.show()
```

```python
#Choose two variables in the data and create a scatterplot.
#decided to use cyl for x and mpg for y
plt.scatter(df['cyl'], df['mpg'])
plt.xlabel('Cylinder Count')
plt.ylabel('Miles per Gallon')
plt.title('Scatterplot of Cylinder Count vs MPG')
plt.show()
#End part A
```
